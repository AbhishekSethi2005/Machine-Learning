```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('car data.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
plt.bar(df['Car_Name'],df['Selling_Price'], color='blue')
plt.title('Selling Price of Cars')
plt.xlabel('Car Name')
plt.ylabel('Selling Price (in Lakhs)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
fuel_count = df['Fuel_Type'].value_counts()
plt.pie(fuel_count,labels=fuel_count.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Fuel Types')
plt.axis('equal')
plt.show()
```

```python
plt.hist(df['Kms_Driven'],bins=5, color = "orange",edgecolor='Black')
plt.title("Distribution of Kms Driven")
plt.xlabel("Kms Driven")
plt.ylabel("No. of Cars")
plt.show()
```
